```python
# importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')


#loading the dataset
df = pd.read_csv(r"D:\pyton folders\datasets\archive\car data.csv")
```

```python
# viewing first row of the data
df.head()
```

```python
# splitting the data into predcitor varaiables and  the target variable
X = df.drop('Selling_Price',axis=1)
X.head()
```

```python
y = df[['Selling_Price']]
y.head()
```

```python
# checking for the categorical values in each categorical column
print(df['Car_Name'].unique(), 2*'\n')
print(df['Fuel_Type'].unique(), '\n')
print(df['Seller_Type'].unique(), '\n')
print(df['Transmission'].unique(), '\n')
```

```python
# extracting the object datatype/categorical columns and putting them in a list
cat_columns = df.select_dtypes(include='object').columns.to_list()
cat_columns
```

```python
# applying one hot encoding to the categorical columns
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(sparse_output=False)
transformed_cols = onehotencoder.fit_transform(df[cat_columns])
transformed_cols
```

```python
#creating a dataframe out of the encoded columns
new_cols = pd.DataFrame(transformed_cols,columns =onehotencoder.get_feature_names_out(cat_columns))
new_cols
```

```python
# concatenating the new dataframe with the original dataframe
new_df = pd.concat([df, new_cols], axis=1)
new_df
```

```python
# dropping the old categorical columns 
dframe = new_df.drop(cat_columns, axis=1)
dframe
```

```python
# putting the new dataframe into the orginal dataframe
df = dframe
df
```

```python
# creating a new X split now 
X = df.drop('Selling_Price', axis=1)
X.head()
```

```python
# checking the shape of the new X split
X.shape
```

```python
# creating a  new y split
y = df[['Selling_Price']]
y.head()
```

```python
# checking the shape of the y split
y.shape
```

```python
# importing linear regression from linear_model class of the sklearn library 
from sklearn.linear_model import LinearRegression

# importing train test split from model_selection class of the sklearn library
from sklearn.model_selection import train_test_split
```

```python
# spliting the data into train and test sets and fitting the data
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=20)
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
# making predictions on the test set
y_pred = linreg.predict(X_test)
y_pred[:33]
```

```python
# importing evaluation metrics to evaluate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score 
print("mean Squared Error: ", mean_squared_error(y_pred, y_test), '\n')
print("Mean Absolute Error: ", mean_absolute_error(y_pred, y_test), '\n')
print("R2 Score: ", r2_score(y_pred, y_test),'\n')
print("Root Mean Squared Error: ", np.mean(mean_absolute_error(y_pred, y_test)))
```
